```python
x = Symbol('x')
f1 = -1/6 * x**4 + 7/6 * x**2
f1
```

```python
def chet_nechet(func):
    if func.subs(x, 1) == func.subs(x, -1):
        print("четная")
    elif func.subs(x, 1) == - func.subs(x, -1):
        print("нечетная")
    else:
        print("функция общего вида")
```

```python
#function 1

# zeros of function
print('нули функции: ', solve(f1))

#chet_nechet
print('четность:')
chet_nechet(f1)

#constancy intervals
print('промежутки знакопостоянства:')
solve_univariate_inequality(f1>0, x)
```
